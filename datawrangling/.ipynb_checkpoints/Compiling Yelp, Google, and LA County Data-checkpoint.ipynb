```python
import pandas as pd
import requests
import numpy as np
import os
```

```python
# Connect yelp data and google data to main data frame

main_df = pd.read_csv('../data/main_df.csv')
# Join Yelp Using program_name and facility_address
yelp_df = pd.read_csv('../data/compiled_yelp_data.csv')
# Join google df to address df then use the facility_address and facility_zip as the columns for joining
google_df = pd.read_csv('../data/geocode_address.csv')
address_df = pd.read_csv('../data/address.csv')
```

```python
compiled_df = pd.merge(main_df, yelp_df, how = 'left', on = ['program_name','facility_address'])
compiled_df = compiled_df.drop(['facility_city_y','hours','Unnamed: 0.1','Unnamed: 0','latitude_yelp','longitude_yelp'], axis = 1)
```

```python
geocode_df = pd.merge(address_df, google_df, how = 'left', on = 'concat_address')
geocode_df = geocode_df.drop(['Unnamed: 0', 'accuracy','google_place_id','status','postcode','concat_address'], axis = 1)
```

```python
compiled_df = pd.merge(compiled_df,geocode_df, how = 'left', on = ['facility_address','facility_zip'])
```

```python
compiled_df = compiled_df.drop(['facility_city','facility_state_y','yelp_id','address'],axis = 1)
```
